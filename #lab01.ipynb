```python
# Importação das bibliotecas necessárias
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt

# Exibir gráficos inline
%matplotlib inline
```

```python
# Carregar imagem de exemplo
from PIL import Image
import requests

url = "https://images.unsplash.com/photo-1746254774131-f94b3d6b84d6?w=500&auto=format&fit=crop&q=60&ixlib=rb-4.1.0&ixid=M3wxMjA3fDB8MHxmZWF0dXJlZC1waG90b3MtZmVlZHwxMjJ8fHxlbnwwfHx8fHw%3D"
image = Image.open(requests.get(url, stream=True).raw)
image = image.convert("RGB")  # Garante 3 canais

# Mostrar imagem
plt.imshow(image)
# plt.title("Imagem Original")
plt.axis('off')
plt.show()
```

Exibir apenas o canal vermelho da imagem

```python
r, g, b = image.split()

fig, axs = plt.subplots(1, figsize=(15, 5))
axs.imshow(r, cmap='grey')
axs.set_title('Red')
axs.axis('off')
plt.show()
```

Modifique aleatoriamente o canal vermelho e gere a imagem novamente

```python
r, g, b = image.split()

fig, axs = plt.subplots(1, figsize=(15, 5))
axs.imshow(r, cmap='Reds')
axs.set_title('Red')
axs.axis('off')
plt.show()
```

Criar uma imagem negativa (inverter as cores)

```python
image = Image.open(requests.get(url, stream=True).raw)
image_inverted = np.array(image)
image_inverted = 255 - image_inverted
plt.imshow(image_inverted)
plt.axis('off')
plt.show()
```

Salvar uma nova imagem modificada no disco

```python
image_modified = Image.fromarray(image_inverted)
image_modified.save("image_modified.jpg")
```

Recorte um pedaço da imagem e salve em disco

```python
image_cut = image.crop((0, 0, 100, 100))
image_cut.save("image_cut.jpg")
```
